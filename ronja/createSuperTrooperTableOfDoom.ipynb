```python
import pandas as pd

# Load the first CSV file
df1 = pd.read_csv('C:\\Users\\Ronja\\DataspellProjects\\BI_Projekt2.0\\data\\data4SuperDuperTableOfDoom.csv')

# Load the second CSV file
df2 = pd.read_csv('C:\\Users\\Ronja\\DataspellProjects\\BI_Projekt2.0\\data\\stellplaetze_2301_2404.csv')

# Combine the two DataFrames
combined_df = df1.combine_first(df2)

# Select only the unique columns
combined_df = combined_df.loc[:,~combined_df.columns.duplicated()]

# Append the columns of df2 that are not in df1 to the list of df1 columns
columns = df1.columns.tolist() + [col for col in df2.columns if col not in df1.columns]

# Reindex the columns of the combined DataFrame to match the new order
combined_df = combined_df.reindex(columns=columns)

# Save the combined DataFrame to a new CSV file
combined_df.to_csv('data4SuperDuperTableOfDoomV2.csv', index=False)
```
