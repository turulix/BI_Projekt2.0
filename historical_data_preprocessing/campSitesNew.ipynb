```python
import pandas as pd


df_raw = pd.read_csv('C:\\Users\\Ronja\\Documents\\Studium\\Semester 4\\DABI2\\Projekt\\Umsatzdaten\\DESTATIS\\Output\\2301_2404\\campSites.csv')

df_cut = df_raw.drop(['Statistik', 'Gebiet', 'Anzahl_Campingplaetze_geoeffnet_Veraenderung_zum_Vorjahresmonat', 'Maximales_Angebot_an_Stellplaetzen_der_letzten_13_Monate'], axis=1)

df_cut = df_cut.loc[df_cut['Land'] != 'Deutschland insgesamt']

month_map = {
    'Januar': '01', 'Februar': '02', 'März': '03', 'April': '04', 'Mai': '05', 'Juni': '06', 'Juli': '07', 'August': '08', 'September': '09', 'Oktober': '10', 'November': '11', 'Dezember': '12'
}
df_cut['Monat'] = df_cut['Monat'].replace(month_map)

df_cut.rename(columns={'Anzahl_Campingplaetze_gesamt': 'campingplaetze_anzahl',
                       'Anzahl_Campingplaetze_geoeffnet': 'urlaubs_campingplaetze_offen',
                       'Angebotene_Stellplaetze_gesamt': 'urlaubs_stellplaetze_anzahl',
                       'Angebotene_Stellplaetze_Veraenderung_zum_Vorjahresmonat': 'change_urlaubs_stellplaetze_offen_vorjahresmonat',
                       'Anteil_der_aktuell_angebotenen_Stellplaetze_am_Maximum': 'anteil_urlaubs_stellplaetze_offen_an_urlaubs_stellplaetze_anzahl'}, inplace=True)

df_cut['date'] = df_cut['Jahr'].astype(str) + '-' + df_cut['Monat'] + '-01'

df_cut = df_cut.drop(['Jahr', 'Monat'], axis=1)

df_cut = df_cut[['Land', 'date', 'campingplaetze_anzahl', 'urlaubs_campingplaetze_offen', 'urlaubs_stellplaetze_anzahl', 'change_urlaubs_stellplaetze_offen_vorjahresmonat', 'anteil_urlaubs_stellplaetze_offen_an_urlaubs_stellplaetze_anzahl']]

int_columns = ['campingplaetze_anzahl', 'urlaubs_campingplaetze_offen', 'urlaubs_stellplaetze_anzahl']
float_columns = ['change_urlaubs_stellplaetze_offen_vorjahresmonat', 'anteil_urlaubs_stellplaetze_offen_an_urlaubs_stellplaetze_anzahl']
df_cut[int_columns] = df_cut[int_columns].apply(pd.to_numeric, errors='coerce').fillna('.')
df_cut[float_columns] = df_cut[float_columns].apply(pd.to_numeric, errors='coerce').fillna('.')

#df_cut.to_csv('C:\\Users\\Ronja\\Documents\\Studium\\Semester 4\\DABI2\\Projekt\\Umsatzdaten\\DESTATIS\\Output\\stellplaetze_2301_2404.csv', index=False)
df_cut
```
